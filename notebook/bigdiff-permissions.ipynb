```python
h  = HTML("<h1>What is the percentage of apps that did not change any permission?</h1>")
h
```

```python
no_changes_df = df_total_perm_changes[df_total_perm_changes.total_permissions_changes == 0]
no_changes_rate = len(no_changes_df) / len(df_total_perm_changes) * 100
print(str(len(no_changes_df)) + " apps out of " + str(len(df_total_perm_changes)) + " ({0:.2f}%)".format(no_changes_rate) + 
      " of the apps do not change the number of used permissions.")
```

```python
h  = HTML("<h1>What is the percentage of apps that removed used permissions?</h1>")
h
```

```python
removed_permissions_df = df_total_perm_changes[df_total_perm_changes.total_permissions_changes < 0]
removed_rate = len(removed_permissions_df) / len(df_total_perm_changes) * 100
print(str(len(removed_changes_df)) + " apps out of " + str(len(df_total_perm_changes)) + " ({0:.2f}%)".format(removed_rate) + 
      " of the apps removed used permissions in their subsequent versions.")
```

```python
h  = HTML("<h1>What is the percentage of apps that added more permissions?</h1>")
h
```

```python
added_permissions_df = df_total_perm_changes[df_total_perm_changes.total_permissions_changes > 0]
added_rate = len(added_permissions_df) / len(df_total_perm_changes) * 100
print(str(len(added_permissions_df)) + " apps out of " + str(len(df_total_perm_changes)) + " ({0:.2f}%)".format(added_rate) + 
      " of the apps added more permissions.")
```

```python
h = HTML("<h1>What is the average number of permissions added between two versions?</h1>")
h
```

```python
# Get apps with two versions
added_perm_with_two_versions = added_permissions_df[added_permissions_df.number_of_versions ==2]
```

```python
print(average(added_perm_with_two_versions.total_permissions_changes))
added_perm_with_two_versions.total_permissions_changes.describe(percentiles=[0.25,0.50,0.75,0.90])
```

```python
h = HTML("<h1>What is the average number of permissions dropped between two versions?</h1>")
h
```

```python
# Get apps with two versions
removed_perm_with_two_versions = removed_permissions_df[removed_permissions_df.number_of_versions == 2 ]
```

```python
removed_perm_with_two_versions_bas = (removed_perm_with_two_versions.total_permissions_changes * -1)
print(average(removed_perm_with_two_versions_bas))
removed_perm_with_two_versions_bas.describe(percentiles=[0.25,0.50,0.75,0.90])
```

```python
h = HTML("<h1>What are the apps with only two versions that have the most number of permission added?")
h
```

```python
result_added_df = added_perm_with_two_versions[added_perm_with_two_versions.total_permissions_changes > 10]

result_added_df.reset_index(inplace=True)
original_df = df.reset_index()
original_df.set_index(keys=['package'], inplace=True)
result_added_df.set_index(keys=['package'], inplace=True)
m_df = pd.merge(result_added_df, original_df, left_index= True, right_index= True)
for package in m_df.index.get_values():
    verc_df = m_df.loc[package, 'version_code']
    verc_int_list = verc_df.tolist()
    verc_list = [str(v) for v in verc_int_list]
    m_df.ix[package, 'version_code_pairs'] = ','.join(verc_list)
m_df.reset_index(inplace=True)
m_df.drop_duplicates(cols=['package'], inplace= True)
cols = ['package','version_code_pairs', 'total_permissions_changes']
m_df = m_df[cols]
m_df = m_df.set_index(keys=['package'])
m_df.sort(columns='total_permissions_changes', ascending=False, inplace=True)
m_df.to_html('../results/bigdiff/out/html/ten_added_permissions.html')
m_df.head()
```

```python
# Detailed view
result_added_df.reset_index(inplace=True)
original_df = df.reset_index()
original_df.set_index(keys=['package'], inplace=True)
result_added_df.set_index(keys=['package'], inplace=True)
pd.merge(result_added_df, original_df, left_index= True, right_index= True).head()
```

```python
h = HTML("<h1>What are the apps with only two versions that have the most number of permission removed?")
h
```

```python
result_removed_df = removed_perm_with_two_versions[removed_perm_with_two_versions.total_permissions_changes < -10]

# Detailed list 1
result_removed_df.reset_index(inplace=True)
original_df = df.reset_index()
original_df.set_index(keys=['package'], inplace=True)
result_removed_df.set_index(keys=['package'], inplace=True)
m_df = pd.merge(result_removed_df, original_df, left_index= True, right_index= True)
for package in m_df.index.get_values():
    verc_df = m_df.loc[package, 'version_code']
    verc_int_list = verc_df.tolist()
    verc_list = [str(v) for v in verc_int_list]
    m_df.ix[package, 'version_code_pairs'] = ','.join(verc_list)
m_df.reset_index(inplace=True)
m_df.drop_duplicates(cols=['package'], inplace= True)
cols = ['package','version_code_pairs', 'total_permissions_changes']
m_df = m_df[cols]
m_df = m_df.set_index(keys=['package'])
m_df.sort(columns='total_permissions_changes', inplace=True)
m_df.to_html('../results/bigdiff/out/html/ten_removed_permissions.html')
m_df.head()
```

```python
# Detailed view
result_removed_df.reset_index(inplace=True)
original_df = df.reset_index()
original_df.set_index(keys=['package'], inplace=True)
result_removed_df.set_index(keys=['package'], inplace=True)
pd.merge(result_removed_df, original_df, left_index= True, right_index= True).head()
```
