```python
import pandas as pd
from IPython.display import HTML
```

```python
h = HTML('<h1>Languages</h1><hr/>'); h
```

```python
df = pd.read_csv('/Users/khalid/uiver/100-apps-2-vers/csv/languages.csv')
df.rename(columns={'Package Name': 'package', 'Version Code':'verc',
                         'Number of Supported Languages':'total_lang', 'Supported Languages List':'lang_list'}, inplace=True)
df.sort(columns=['package', 'verc'], inplace=True)
df.set_index(['package', 'verc'], inplace =True)
#df.to_excel('/Users/khalid/git/appsmining/results/tmp/languages.xls')
df
```

```python
def get_langugae_changes(dataframe):
    rdf = pd.DataFrame(columns=['package_name', 'versions_count', 'language_count_changes'])
    rdf.set_index('package_name', inplace =True)
    for gp in dataframe.index.get_level_values('package').unique():
        tmp_df = dataframe.loc[gp,'total_lang']
        d = 0
        if len(tmp_df) == 1:
            d = 0
        else:
            for r in tmp_df:
                d = r - d
        rdf.ix[gp, 'language_count_changes'] = d
        rdf.ix[gp, 'versions_count'] = len(tmp_df)
    return rdf
lang_df = get_langugae_changes(df)
plt.title('Changes in supported languages')
xlabel('changes of languages')
ylabel('number of apps')
lang_df.language_count_changes.hist(figsize=(4, 4), bins=20)
savefig('/Users/khalid/git/appsmining/results/images/lang_changes.png', dpi=200, format='png')
lang_df
```

```python
print ('min => ' + str(lang_df.language_count_changes.min()))
print ('max => ' + str(lang_df.language_count_changes.max()))
print ('mean => ' + str(lang_df.language_count_changes.mean()))
print ('std => ' + str(lang_df.language_count_changes.std()))
```

```python
df = pd.read_csv('/Users/khalid/uiver/100-apps-2-vers/csv/screen_sizes.csv')
df.rename(columns={'Package Name': 'package', 'Version Code':'verc',
                         'Number of Supported Screen Sizes':'total_screens'}, inplace=True)
df.sort(columns=['package', 'verc'], inplace=True)
df.set_index(['package', 'verc'], inplace =True)
df.to_excel('/Users/khalid/git/appsmining/results/tmp/screen_sizes.xls')
df
```

```python
def get_screen_size_changes(dataframe):
    rdf = pd.DataFrame(columns=['package_name', 'versions_count', 'screen_sizes_count_changes'])
    rdf.set_index('package_name', inplace =True)
    for gp in dataframe.index.get_level_values('package').unique():
        tmp_df = dataframe.loc[gp,'total_screens']
        d = 0
        if len(tmp_df) == 1:
            d = 0
        else:
            for r in tmp_df:
                d = r - d
        rdf.ix[gp, 'screen_sizes_count_changes'] = d
        rdf.ix[gp, 'versions_count'] = len(tmp_df)
    return rdf
screen_df = get_screen_size_changes(df)
plt.title('Changes in supported screens')
xlabel('changes of supported screens')
ylabel('number of apps')
screen_df.screen_sizes_count_changes.hist(figsize=(4, 4), bins=20)
savefig('/Users/khalid/git/appsmining/results/images/screen_sizes.png', dpi=200, format='png')
screen_df
```

```python
print ('min => ' + str(screen_df.screen_sizes_count_changes.min()))
print ('max => ' + str(screen_df.screen_sizes_count_changes.max()))
print ('mean => ' + str(screen_df.screen_sizes_count_changes.mean()))
print ('std => ' + str(screen_df.screen_sizes_count_changes.std()))
```

```python
df_screen_sizes_sorted = df.reset_index().sort(['package', 'verc'])
df_screen_sizes_no_duplicates = df_screen_sizes_sorted.drop_duplicates(subset='package', take_last=True)

df_small_summary = df_screen_sizes_no_duplicates.groupby('small').count()
df_small_summary.plot(kind='pie', y='package', autopct='%.2f', title='Small Screen Support', figsize=(4, 4), colors=['r', 'g'], fontsize=20)
savefig('/Users/khalid/git/appsmining/results/images/screen_sizes_small.png', dpi=100, format='png')

df_normal_summary = df_screen_sizes_no_duplicates.groupby('normal').count()
df_normal_summary.plot(kind='pie', y='package', autopct='%.2f', title='Normal Screen Support', figsize=(4, 4), colors=['g', 'r'], fontsize=20)
savefig('/Users/khalid/git/appsmining/results/images/screen_sizes_normal.png', dpi=100, format='png')

df_large_summary = df_screen_sizes_no_duplicates.groupby('large').count()
df_large_summary.plot(kind='pie', y='package', autopct='%.2f', title='Large Screen Support', figsize=(4, 4), colors=['g', 'r'], fontsize=20)
savefig('/Users/khalid/git/appsmining/results/images/screen_sizes_large.png', dpi=100, format='png')

df_xlarge_summary = df_screen_sizes_no_duplicates.groupby('xlarge').count()
df_xlarge_summary.plot(kind='pie', y='package', autopct='%.2f', title='XLarge Screen Support', figsize=(4, 4),  colors=['r', 'g'], fontsize=20)
savefig('/Users/khalid/git/appsmining/results/images/screen_sizes_xlarge.png', dpi=100, format='png')
```
